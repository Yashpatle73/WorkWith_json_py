```python
import json
```

```python
import matplotlib.pyplot as plt
    
```

```python
men=[]
women=[]
with open('medical.json')as fr:
    data=json.load(fr)
    for i in data ['medical']['men']:
        men.append(i['level'])
    for i in data ['medical']['women']:
        women.append(i['level'])
plt.hist([men,women],bins=[80,100,125,150],color=['green','red'],rwidth=.95,label=['men','women'])
plt.legend()
plt.title('BLOOD SUGER CHART')
plt.xlabel('Suger lavel')
plt.ylabel('number')
plt.show()
        
```
